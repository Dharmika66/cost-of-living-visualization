# Cost of Living vs Average Salary â€“ Data Visualization

This project visualizes cost of living and salary data across global cities using a synthetic dataset. Charts explore affordability, regional comparisons, and breakdowns by category.

```python
%pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("cost_of_living_dataset.csv")
df.head()
```

## Average Salary vs Total Monthly Cost (Top 15 Cities)

```python
top15 = df.sort_values("Avg Monthly Salary ($)", ascending=False).head(15)
plt.figure(figsize=(12,6))
bar1 = sns.barplot(data=top15, x="City", y="Avg Monthly Salary ($)", color="green", label="Salary")
bar2 = sns.barplot(data=top15, x="City", y="Total Monthly Cost ($)", color="red", label="Cost")
plt.xticks(rotation=45)
plt.legend()
plt.title("Salary vs Cost of Living by City")
plt.tight_layout()
plt.show()
```

## Affordability Index by City

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df.sort_values("Affordability Index", ascending=False), x="City", y="Affordability Index", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Affordability Index (Salary - Cost)")
plt.tight_layout()
plt.show()
```

## Average Salary and Cost by Region

```python
region_grouped = df.groupby("Region")[["Avg Monthly Salary ($)", "Total Monthly Cost ($)"]].mean().round(2)
region_grouped.plot(kind="bar", figsize=(10,5), colormap="Set2")
plt.title("Average Salary and Cost by Region")
plt.ylabel("USD ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
